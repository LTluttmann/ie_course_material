```python
# Standard imports used throughout the course
import spacy
from rich import print as rprint

from ie_course.utils import ents_to_dict, print_ents
```

## 1. Load a spaCy model

```python
nlp = spacy.load('en_core_web_sm')
print('Model loaded:', nlp.meta['name'], nlp.meta['version'])
```

## 2. Your first IE pipeline

```python
text = (
    'Apple was founded by Steve Jobs, Steve Wozniak, and Ronald Wayne '
    'on April 1, 1976, in Cupertino, California.'
)

doc = nlp(text)
print_ents(doc)
```

```python
# Group entities by label
rprint(ents_to_dict(doc))
```
